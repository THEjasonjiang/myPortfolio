### Ashes Team Batting Innings Data

```python
# data processing
import pandas as pd
# data visualisation libraries
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# read csv file and display 5 random rows 
ashes_team_data = pd.read_csv("ashes_team_innings.csv")
ashes_team_data.sample(n=5)
```

The dataset contains every Ashes batting innings with the following columns/features:

|FEATURE(S)|DESCRIPTION|
|----|----|
|Match ID|Ashes test match ID|
|Runs|Runs scored in innings by batting team|
|Wickets|Wickets taken in innings by bowling team|
|Overs|Overs bowled by bowling team|
|Balls|Balls bowled by bowling team|
|BPO|Number of balls bowled per over|
|Run Rate|Runs scored per over|
|Minutes|Time taken in innings|
|Total Extras|Runs scored not by batting team|
|No-Balls|Runs when bowler commits illegal action when bowling|
|Wides|Runs when bowler delivers too far from batter to hit|
|Byes|Runs when batters run without striking batter's bat or body|
|Leg-Byes|Runs when batters run when batter is struck on body|
|Penalty Runs|Runs awarded for breaches of cricketing laws|
|Home Team & Home Team ID|Team which is playing on own ground|
|Season & Season ID|Time of year match is played|
|Series Name & Series ID|Name of series match played in|
|Team & Team ID|Country team of innings|
|Town Area, Town Name & Town ID|Geographic location of match played|
|Ground Name & Ground ID|Name of ground match played in|

Batting average is the ratio of a player's number of runs to the number of times they have been out. It is a good metric as it demonstrates the consistency of a batsmen. A high batting average indicates that a batsman is able to consistently score runs without getting out frequently. In addition, it allows for easy comparison between different batsmen and helps in assessing who has performed better in terms of runs scored per dismissal.

The grouped bar chart plot shows that **David Gower**, **Steve Waugh**, and **Steven Smith** all follow a trend of performing well at the start of an Ashes test match and get gradually worse as a test match goes on. This means their run averages decrease when the match innings increase.

The main reason for this trend is due to 'Pitch Wear and Tear'. Pitch Wear and Tear refers to the changes that occur on the pitch over the course of a match and this especially happens during test matches since they are 5 day matches. A pitch undergoes changes that can impact the behaviour of the ball, influencing how it bounces and spins, hence favouring bowlers. The primary factors contributing to pitch wear and tear include:

* Bounce and Pace Changes: Initially, a fresh pitch may offer more consistent bounce and pace to fast bowlers. However, as the match progresses, the surface can wear down, leading to a decrease in pace and a potentially lower and uneven bounce giving bowlers advantages.

* Cracks and Roughness: With time, especially in hot and dry conditions, cracks may develop on the pitch. These cracks can affect the ball's trajectory and provide additional challenges for batsmen. Bowlers may exploit the rough areas to generate more spin or movement.

* Footmarks: Bowlers leave footmarks on the pitch as they deliver the ball. Over time, these footmarks can become areas of uneven wear, creating more challenges for batsmen. Spin bowlers, in particular, may target these areas to extract extra turn.

Because of these factors, it's surprising to see **Graham Gooch's** highest run average is in the 4th innings of a match and every other batsmen's performances in the 4th innings of a match on average is their worst.

Another possible factor of low 4th match innings run averages is some matches aren't going long enough to get to a 4th innings. Some matches could finish before the 4th innings and so some batsmen might not have enough chances to bat during the 4th innings of a test match.

```python
plt.figure(figsize=(15,10))
colors = ['red','skyblue','black','pink','green','brown']
for i,name in enumerate(ashes_player_bowling_data['name'].unique()):
    cumulative_wkts = [0]+[sum(ashes_player_bowling_data[ashes_player_bowling_data['name']==name]['wickets'][:j+1]) for j in range(len(ashes_player_bowling_data[ashes_player_bowling_data['name']==name]['wickets']))]
    plt.plot([0]+list(ashes_player_bowling_data[ashes_player_bowling_data['name']==name].reset_index().index+1), cumulative_wkts, linestyle='-', color=colors[i])
plt.title('Cumulative Wickets Over Innings')
plt.xlabel('Innings')
plt.ylabel('Cumulative Wickets')
plt.grid(True)
plt.legend(handles=[plt.Line2D([0], [0], linewidth=2, color=color, label=label) for label, color in zip(ashes_player_bowling_data['name'].unique(), colors)], loc='upper left', fontsize='x-large')
plt.show()
```
